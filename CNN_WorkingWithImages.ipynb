# Converting Img to GrayScale Img

```python
plt.imshow(img)
```

```python
R = img[:,:,0]
G = img[:,:,1]
B = img[:,:,2]
```

```python
# Uniform Averaging
# (0.333*R + 0.333*G + 0.333*B )

imgGrey = 0.333*R + 0.333*G + 0.333*B
plt.imshow(imgGrey, cmap='gray')
```

```python
# Grenrating more accurate GrayScale img

imgGrey2 = 0.2989*R + 0.5870*G + 0.1140*B
plt.imshow(imgGrey2, cmap='grey')
```
